# But what is Euler's number, $e$?

The number <wiki:e_(mathematical_constant)> is a mathematical constant, approximately equal to $2.71828$, that is the base of the natural logarithm and exponential function. It is sometimes called **Euler's number**, after the Swiss mathematician <wiki:Leonhard_Euler>, or alternatively **Napier's constant** after the Scottish mathematician <wiki:John_Napier>. The Swiss mathematician <wiki:Jacob_Bernoulli> discovered the constant while studying compound interest.

The constant $e$ was introduced by <wiki:Jacob_Bernoulli> in 1683, for solving the problem of continuous compounding of interest. In his solution, the constant $e$ occurs as the limit:

$$
\lim_{n \rightarrow \infty} \left( 1 + \frac{1}{n} \right)^n
$$

```python
# python code
import math

def myE(n):
    # return ((1+1/n)**n)
    return ((2*n+1)/(2*n-1))**n
```

```python
myE(100000)
```

```python
# python code
import math

def exp(x):
    # return 1 + x + x**2 / 2 + x**3 /6 + ...
    return sum([
        x**n / math.factorial(n)
        for n in range(0, 100)
    ])
```

```python
exp(1)
```
